```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import numpy as np
import os
import time
```

```python
# Create a folder to save all outputs
output_dir = "mlp_results"
os.makedirs(output_dir, exist_ok=True)
```

```python
iris = load_iris()
X = iris.data # Features
y = iris.target # Labels

print("First 5 rows of features (X):")
print(X[:5])
print("\nFirst 5 rows of targets (y):")
print(y[:5])
print("\nX shape:", X.shape)
print("y shape:", y.shape)
# Split the data into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#CODE A
# ============================================================
# TASK 1: Search for best hidden layer configuration
# ============================================================

# Define all configurations to test
# Each tuple = one hidden layer per value, value = neurons in that layer
# Example: (10,) = 1 hidden layer with 10 neurons
#          (10, 10) = 2 hidden layers with 10 neurons each
#          (50, 30, 10) = 3 hidden layers with 50, 30, 10 neurons
configurations = [
    (1,),
    (2,),
    (3,),
    (5,),
    (10,),
    (20,),
    (50,),
    (100,),
    (10, 10),
    (20, 20),
    (50, 50),
    (100, 100),
    (50, 30, 10),
    (100, 50, 25),
]

task1_accuracies = []      # stores accuracy for each config
task1_epochs = []          # stores how many epochs each config needed
task1_loss_curves = []     # stores full loss curve for each config
task1_labels = []          # stores readable label like "(10,)" for plots
task1_times = []           # tracks time

print("=" * 60)
print("TASK 1: Testing different hidden layer configurations")
print("=" * 60)

for config in configurations:

    # Build and train a model for this configuration
    # learning_rate_init is fixed at 0.001 for Task 1
    # We change ONLY the hidden_layer_sizes
    mlp_temp = MLPClassifier(
        hidden_layer_sizes=config,
        max_iter=1000,
        random_state=42,
        learning_rate_init=0.001
    )
    start_time = time.time()               # NEW: start timer
    mlp_temp.fit(X_train_scaled, y_train)
    elapsed = time.time() - start_time     # NEW: stop timer

    # Predict and measure accuracy
    y_pred_temp = mlp_temp.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred_temp)

    # Store results
    task1_accuracies.append(acc)
    task1_epochs.append(mlp_temp.n_iter_)
    task1_loss_curves.append(mlp_temp.loss_curve_)
    task1_labels.append(str(config))
    task1_times.append(elapsed)            # NEW: store time

    print(f"Config {str(config):20s} | Accuracy: {acc:.4f} | Epochs: {mlp_temp.n_iter_} | Time: {elapsed:.3f}s")

# Find the best configuration (highest accuracy)
best_config_index = int(np.argmax(task1_accuracies))
best_config = configurations[best_config_index]
best_config_accuracy = task1_accuracies[best_config_index]

print(f"\nBest Configuration: {best_config}")
print(f"Best Accuracy:      {best_config_accuracy:.4f}")

# ---- PLOT 1: Accuracy bar chart for all configurations ----
fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.bar(task1_labels, task1_accuracies, color='steelblue', edgecolor='black')
bars[best_config_index].set_color('orange')   # highlight best in orange
ax.set_title('Task 1: Accuracy per Hidden Layer Configuration', fontsize=14)
ax.set_xlabel('Hidden Layer Configuration')
ax.set_ylabel('Accuracy')
ax.set_ylim([0, 1.1])
ax.tick_params(axis='x', rotation=45)
# Add value labels on top of each bar
for bar, acc in zip(bars, task1_accuracies):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.01,
            f'{acc:.3f}',
            ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task1_accuracy_comparison.png"), dpi=150)
plt.show()
print("Saved: task1_accuracy_comparison.png")

# ---- PLOT 2: All loss curves on one graph ----
plt.figure(figsize=(12, 7))
for i, (curve, label) in enumerate(zip(task1_loss_curves, task1_labels)):
    linewidth = 3 if i == best_config_index else 1
    linestyle = '-' if i == best_config_index else '--'
    plt.plot(curve, label=label, linewidth=linewidth, linestyle=linestyle)
plt.title('Task 1: Learning Curves for All Configurations\n(Bold line = best config)', fontsize=13)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='upper right', fontsize=8)
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task1_all_loss_curves.png"), dpi=150)
plt.show()
print("Saved: task1_all_loss_curves.png")

# ---- PLOT 3: Epochs needed per configuration ----
fig, ax = plt.subplots(figsize=(12, 5))
bars2 = ax.bar(task1_labels, task1_epochs, color='mediumseagreen', edgecolor='black')
bars2[best_config_index].set_color('orange')
ax.set_title('Task 1: Epochs to Converge per Configuration', fontsize=14)
ax.set_xlabel('Hidden Layer Configuration')
ax.set_ylabel('Number of Epochs')
ax.tick_params(axis='x', rotation=45)
for bar, ep in zip(bars2, task1_epochs):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 5,
            str(ep),
            ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task1_epochs_comparison.png"), dpi=150)
plt.show()
print("Saved: task1_epochs_comparison.png")

# ---- PLOT 3b: Training time per configuration ----
fig, ax = plt.subplots(figsize=(12, 5))
bars_t1 = ax.bar(task1_labels, task1_times, color='salmon', edgecolor='black')
bars_t1[best_config_index].set_color('orange')
ax.set_title('Task 1: Training Time per Configuration (seconds)', fontsize=14)
ax.set_xlabel('Hidden Layer Configuration')
ax.set_ylabel('Time (seconds)')
ax.tick_params(axis='x', rotation=45)
for bar, t in zip(bars_t1, task1_times):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.001,
            f'{t:.3f}s',
            ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task1_time_comparison.png"), dpi=150)
plt.show()
print("Saved: task1_time_comparison.png")

# ---- PLOT 4: Loss curve of BEST config only ----
plt.figure(figsize=(8, 5))
plt.plot(task1_loss_curves[best_config_index], color='orange', linewidth=2, label=f'Best Config {best_config}')
plt.title(f'Task 1: Loss Curve of Best Configuration {best_config}', fontsize=13)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task1_best_loss_curve.png"), dpi=150)
plt.show()
print("Saved: task1_best_loss_curve.png")

# ---- Train best config fully and print its classification report ----
mlp_best_config = MLPClassifier(
    hidden_layer_sizes=best_config,
    max_iter=1000,
    random_state=42,
    learning_rate_init=0.001
)
mlp_best_config.fit(X_train_scaled, y_train)
y_pred_best_config = mlp_best_config.predict(X_test_scaled)

print(f"\nDetailed Report for Best Configuration {best_config}:")
print(classification_report(y_test, y_pred_best_config, target_names=iris.target_names))
print(f"Layers: {mlp_best_config.n_layers_} | Activation: {mlp_best_config.activation} | Epochs: {mlp_best_config.n_iter_}")


# ============================================================
# TASK 2: Search for best learning rate (using best config)
# ============================================================

# Define learning rates to test
# We keep best_config fixed and only change learning_rate_init
learning_rates = [0.0001, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5]

task2_accuracies = []
task2_epochs = []
task2_loss_curves = []
task2_labels = []
task2_times = []

print("\n" + "=" * 60)
print(f"TASK 2: Testing learning rates with best config = {best_config}")
print("=" * 60)

for lr in learning_rates:

    mlp_lr = MLPClassifier(
        hidden_layer_sizes=best_config,
        max_iter=1000,
        random_state=42,
        learning_rate_init=lr
    )
    start_time = time.time()               # NEW: start timer
    mlp_lr.fit(X_train_scaled, y_train)
    elapsed = time.time() - start_time     # NEW: stop timer

    y_pred_lr = mlp_lr.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred_lr)

    task2_accuracies.append(acc)
    task2_epochs.append(mlp_lr.n_iter_)
    task2_loss_curves.append(mlp_lr.loss_curve_)
    task2_labels.append(str(lr))
    task2_times.append(elapsed)            # NEW: store time

    print(f"LR: {str(lr):8s} | Accuracy: {acc:.4f} | Epochs: {mlp_lr.n_iter_} | Time: {elapsed:.3f}s")

# Find best learning rate by ACCURACY only
best_lr_acc_index = int(np.argmax(task2_accuracies))

# Find best learning rate by CONVERGENCE SPEED (fewest epochs)
best_lr_speed_index = int(np.argmin(task2_epochs))

# Find best BALANCE between accuracy and convergence speed
# Normalize both metrics to 0-1 scale so they are comparable
acc_array = np.array(task2_accuracies)
epochs_array = np.array(task2_epochs)

acc_normalized = (acc_array - acc_array.min()) / (acc_array.max() - acc_array.min() + 1e-9)
# For epochs: FEWER is better, so we invert it
speed_normalized = 1 - (epochs_array - epochs_array.min()) / (epochs_array.max() - epochs_array.min() + 1e-9)

# Balance score = equal weight to accuracy and speed
balance_scores = 0.5 * acc_normalized + 0.5 * speed_normalized

best_lr_index = int(np.argmax(balance_scores))   # THIS is used for all plots
best_lr = learning_rates[best_lr_index]
best_lr_accuracy = task2_accuracies[best_lr_index]

print(f"\n--- Learning Rate Analysis ---")
print(f"Best by Accuracy only       : LR={learning_rates[best_lr_acc_index]} | Acc={task2_accuracies[best_lr_acc_index]:.4f} | Epochs={task2_epochs[best_lr_acc_index]}")
print(f"Best by Speed only          : LR={learning_rates[best_lr_speed_index]} | Acc={task2_accuracies[best_lr_speed_index]:.4f} | Epochs={task2_epochs[best_lr_speed_index]}")
print(f"Best BALANCE (acc + speed)  : LR={best_lr} | Acc={best_lr_accuracy:.4f} | Epochs={task2_epochs[best_lr_index]}")

# ---- PLOT 5: Accuracy per learning rate ----
fig, ax = plt.subplots(figsize=(10, 5))
bars3 = ax.bar(task2_labels, task2_accuracies, color='cornflowerblue', edgecolor='black')
bars3[best_lr_index].set_color('tomato')
ax.set_title(f'Task 2: Accuracy per Learning Rate\n(Config fixed = {best_config})', fontsize=13)
ax.set_xlabel('Learning Rate')
ax.set_ylabel('Accuracy')
ax.set_ylim([0, 1.1])
for bar, acc in zip(bars3, task2_accuracies):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.01,
            f'{acc:.3f}',
            ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task2_lr_accuracy_comparison.png"), dpi=150)
plt.show()
print("Saved: task2_lr_accuracy_comparison.png")

# ---- PLOT 6: All loss curves per learning rate ----
plt.figure(figsize=(12, 7))
for i, (curve, label) in enumerate(zip(task2_loss_curves, task2_labels)):
    linewidth = 3 if i == best_lr_index else 1
    linestyle = '-' if i == best_lr_index else '--'
    plt.plot(curve, label=f'LR={label}', linewidth=linewidth, linestyle=linestyle)
plt.title(f'Task 2: Loss Curves per Learning Rate\n(Bold = best LR={best_lr})', fontsize=13)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='upper right', fontsize=9)
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task2_all_lr_loss_curves.png"), dpi=150)
plt.show()
print("Saved: task2_all_lr_loss_curves.png")

# ---- PLOT 7: Epochs per learning rate ----
fig, ax = plt.subplots(figsize=(10, 5))
bars4 = ax.bar(task2_labels, task2_epochs, color='plum', edgecolor='black')
bars4[best_lr_index].set_color('tomato')
ax.set_title(f'Task 2: Epochs to Converge per Learning Rate\n(Config fixed = {best_config})', fontsize=13)
ax.set_xlabel('Learning Rate')
ax.set_ylabel('Number of Epochs')
for bar, ep in zip(bars4, task2_epochs):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 3,
            str(ep),
            ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task2_lr_epochs_comparison.png"), dpi=150)
plt.show()
print("Saved: task2_lr_epochs_comparison.png")

# ---- PLOT 7b: Balance Score per learning rate ----
fig, ax = plt.subplots(figsize=(10, 5))
bars_bal = ax.bar(task2_labels, balance_scores, color='mediumpurple', edgecolor='black')
bars_bal[best_lr_index].set_color('gold')
# Also mark the pure-accuracy best and pure-speed best differently
bars_bal[best_lr_acc_index].set_edgecolor('tomato')
bars_bal[best_lr_acc_index].set_linewidth(3)
bars_bal[best_lr_speed_index].set_edgecolor('deepskyblue')
bars_bal[best_lr_speed_index].set_linewidth(3)

ax.set_title(
    f'Task 2: Balance Score per Learning Rate\n'
    f'Gold=Best Balance | Red border=Best Accuracy | Blue border=Fastest',
    fontsize=12
)
ax.set_xlabel('Learning Rate')
ax.set_ylabel('Balance Score (higher = better)')
ax.set_ylim([0, 1.2])
for bar, score in zip(bars_bal, balance_scores):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.02,
            f'{score:.3f}',
            ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task2_balance_score.png"), dpi=150)
plt.show()
print("Saved: task2_balance_score.png")

# ---- PLOT 7c: Training time per learning rate ----
fig, ax = plt.subplots(figsize=(10, 5))
bars_t2 = ax.bar(task2_labels, task2_times, color='lightsalmon', edgecolor='black')
bars_t2[best_lr_index].set_color('tomato')
ax.set_title(f'Task 2: Training Time per Learning Rate (seconds)\n(Config fixed = {best_config})', fontsize=13)
ax.set_xlabel('Learning Rate')
ax.set_ylabel('Time (seconds)')
for bar, t in zip(bars_t2, task2_times):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 0.001,
            f'{t:.3f}s',
            ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task2_time_comparison.png"), dpi=150)
plt.show()
print("Saved: task2_time_comparison.png")

# ---- PLOT 8: Best LR loss curve only ----
plt.figure(figsize=(8, 5))
plt.plot(task2_loss_curves[best_lr_index], color='tomato', linewidth=2, label=f'Best LR={best_lr}')
plt.title(f'Task 2: Loss Curve of Best Learning Rate = {best_lr}', fontsize=13)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(output_dir, "task2_best_lr_loss_curve.png"), dpi=150)
plt.show()
print("Saved: task2_best_lr_loss_curve.png")

# ---- Train final best model and print full report ----
mlp_final = MLPClassifier(
    hidden_layer_sizes=best_config,
    max_iter=1000,
    random_state=42,
    learning_rate_init=best_lr
)
mlp_final.fit(X_train_scaled, y_train)
y_pred_final = mlp_final.predict(X_test_scaled)

print("\n" + "=" * 60)
print("FINAL MODEL (Best Config + Best Learning Rate)")
print("=" * 60)
print(f"Hidden Layer Config : {best_config}")
print(f"Learning Rate       : {best_lr}")
print(f"Accuracy            : {accuracy_score(y_test, y_pred_final):.4f}")
print(f"Epochs              : {mlp_final.n_iter_}")
print(f"Layers              : {mlp_final.n_layers_}")
print(f"Activation          : {mlp_final.activation}")
print(f"Output Activation   : {mlp_final.out_activation_}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_final, target_names=iris.target_names))


# ============================================================
# SAVE SUMMARY TEXT FILE
# ============================================================

summary_path = os.path.join(output_dir, "summary.txt")
with open(summary_path, "w") as f:

    f.write("=" * 60 + "\n")
    f.write("TASK 1: Hidden Layer Configuration Search\n")
    f.write("=" * 60 + "\n")
    f.write(f"{'Configuration':<25} {'Accuracy':>10} {'Epochs':>10} {'Time(s)':>10}\n")
    f.write("-" * 62 + "\n")
    for i, config in enumerate(configurations):
        marker = " <-- BEST" if i == best_config_index else ""
        f.write(f"{str(config):<25} {task1_accuracies[i]:>10.4f} {task1_epochs[i]:>10} {task1_times[i]:>10.3f}{marker}\n")

    f.write("\n" + "=" * 60 + "\n")
    f.write("TASK 2: Learning Rate Search\n")
    f.write(f"(Using best config = {best_config})\n")
    f.write("=" * 60 + "\n")
    f.write(f"{'Learning Rate':<20} {'Accuracy':>10} {'Epochs':>10} {'Time(s)':>10} {'Balance':>10}\n")
    f.write("-" * 70 + "\n")
    for i, lr in enumerate(learning_rates):
        markers = []
        if i == best_lr_index:       markers.append("BEST BALANCE")
        if i == best_lr_acc_index:   markers.append("BEST ACCURACY")
        if i == best_lr_speed_index: markers.append("FASTEST")
        marker_str = " <-- " + ", ".join(markers) if markers else ""
        f.write(f"{str(lr):<20} {task2_accuracies[i]:>10.4f} {task2_epochs[i]:>10} {task2_times[i]:>10.3f} {balance_scores[i]:>10.3f}{marker_str}\n")

    f.write(f"\nBest by Accuracy : LR={learning_rates[best_lr_acc_index]} | Acc={task2_accuracies[best_lr_acc_index]:.4f} | Epochs={task2_epochs[best_lr_acc_index]}\n")
    f.write(f"Fastest          : LR={learning_rates[best_lr_speed_index]} | Acc={task2_accuracies[best_lr_speed_index]:.4f} | Epochs={task2_epochs[best_lr_speed_index]}\n")
    f.write(f"Best Balance     : LR={best_lr} | Acc={best_lr_accuracy:.4f} | Epochs={task2_epochs[best_lr_index]}\n")

    f.write("\n" + "=" * 60 + "\n")
    f.write("FINAL MODEL SUMMARY\n")
    f.write("=" * 60 + "\n")
    f.write(f"Best Hidden Config : {best_config}\n")
    f.write(f"Best Learning Rate : {best_lr}\n")
    f.write(f"Final Accuracy     : {accuracy_score(y_test, y_pred_final):.4f}\n")
    f.write(f"Epochs             : {mlp_final.n_iter_}\n")
    f.write(f"Layers             : {mlp_final.n_layers_}\n")
    f.write(f"Activation         : {mlp_final.activation}\n")
    f.write(f"Output Activation  : {mlp_final.out_activation_}\n")
    f.write("\nClassification Report:\n")
    f.write(classification_report(y_test, y_pred_final, target_names=iris.target_names))

print(f"\nSaved: summary.txt")
print(f"\nAll results saved to folder: '{output_dir}/'")

from google.colab import files
import zipfile

# Zip the entire results folder
zip_path = "mlp_results.zip"
with zipfile.ZipFile(zip_path, 'w') as zipf:
    for root, dirs, filenames in os.walk(output_dir):
        for filename in filenames:
            zipf.write(os.path.join(root, filename))

# Trigger automatic download to your computer
files.download(zip_path)
```

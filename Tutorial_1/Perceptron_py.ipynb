```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle

class Perceptron (object):
  def __init__(self, eta=0.01, n_iter=10):
    self.eta = eta
    self.n_iter = n_iter
  def weighted_sum(self, X):
    return np.dot(X, self.w_[1:]) + self.w_[0]
    #Task 2:
  #ORIGINAL STEP FUNCTION
  #def predict(self, X):
  #  return np.where(self.weighted_sum(X) >= 0.0, 1, -1)

  #MODIFIED SIGMOID
  def predict(self, X):
      # 1. Get the weighted sum
      z = self.weighted_sum(X)

      # 2. Apply the Sigmoid mathematical formula
      sigmoid_output = 1.0 / (1.0 + np.exp(-z))

      # 3. Return 1 if probability is >= 0.5, otherwise return 0
      return np.where(sigmoid_output >= 0.5, 1, 0)
  def fit(self, X, y):
    self.w_ = np.zeros(1+X.shape[1])
    self.errors_ = []
    print("Weights:", self.w_)
    for _ in range(self.n_iter):
      errors = 0
      for xi, target in zip(X, y):
        y_pred = self.predict(xi)
        update = self.eta * (target - y_pred)
        self.w_[1:] += update * xi
        print("Updated Weights:", self.w_[1:])
        self.w_[0] += update
        errors += int(update != 0.0)
      self.errors_.append(errors)
    return self
```

```python


df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
df = shuffle(df)
df.head()
```

```python
X = df.iloc[:, 0:4].values
y = df.iloc[:,4].values
print(X[0:5])
print(y[0:5])
from sklearn.model_selection import train_test_split
train_data, test_data, train_labels, test_labels = train_test_split(X, y, test_size=0.3)
train_labels = np.where(train_labels == 'Iris-setosa', 1, 0)
test_labels = np.where(test_labels == 'Iris-setosa', 1, 0)
print('Train data:', train_data[0:2])
print('Train labels:', train_labels[0:2])
print('Test data:', test_data[0:2])
print('Test labels:', test_labels[0:2])
```

```python
#from sklearn.linear_model import Perceptron
perceptron = Perceptron(eta=0.1, n_iter=10)
perceptron.fit(train_data, train_labels)

test_preds = perceptron.predict(test_data)
print(test_preds)
```

```python
from sklearn.metrics import accuracy_score
y_preds = perceptron.predict(test_data)
accuracy = accuracy_score(test_labels, y_preds)
print('Accuracy:', round(accuracy, 2) * 100, '%')
```
